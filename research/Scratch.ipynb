```python
from ebooklib import epub
```

```python
book = epub.read_epub("data/René Girard_ James G. Williams - I See Satan Fall Like Lightning (2012, Orbis Books) - libgen.li.epub")
```

```python
toc = book.toc
```

```python
def get_sections(toc,spaces=0):
    for e in toc:
        if hasattr(e,"__iter__"):
            get_sections(e,spaces=spaces+1)
        else:
            print(" "*spaces + e.title)
```

```python
get_sections(toc)
```
